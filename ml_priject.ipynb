```python
# Step 1: Import necessary libraries.
import pandas as pd
import re
from google.colab import files

# Step 2: Upload your CSV file.
uploaded = files.upload()

# Verify each uploaded file.
for filename in uploaded.keys():
    print('Uploaded file "{name}" with length {length} bytes'.format(
        name=filename, length=len(uploaded[filename])))

# Step 3: Define the sentiment labeling function with error handling.
def label_sentiment(verse):
    """
    Analyze the verse (شعر) as a whole and assign a sentiment label.

    Sentiment Labels (in Urdu):
      - Positive (مثبت)
      - Negative (منفی)
      - Neutral (غیر جانب دار)
      - Mixed (مخلوط)
      - Optimistic (امید افزا)
      - Pessimistic (مایوس)
      - Inspirational (حوصلہ افزا)
      - Philosophical (فلسفیانہ)
    """
    # If the input is not a string (e.g. NaN or numeric), return a neutral sentiment.
    if not isinstance(verse, str):
        return "غیر جانب دار"

    # Normalize text for easier matching.
    text = verse.strip().lower()

    # Rule 1: Check for expressions of hope/inspiration.
    if re.search(r'\bامید\b', text) or re.search(r'\bچراغ\b', text) or re.search(r'\bروشنی\b', text):
        if re.search(r'\bآزمائش\b', text) or re.search(r'\bمشکل\b', text) or re.search(r'\bکانٹ\b', text):
            return "حوصلہ افزا"  # Inspirational: hope amid struggles.
        return "امید افزا"     # Optimistic sentiment.

    # Rule 2: Check for negative emotions.
    if re.search(r'\bدکھ\b', text) or re.search(r'\bغم\b', text) or re.search(r'\bرنج\b', text) or re.search(r'\bافسردہ\b', text):
        return "منفی"

    # Rule 3: Check for sentiments related to love.
    if re.search(r'\bمحبت\b', text):
        if re.search(r'\bجدائی\b', text) or re.search(r'\bتنہائی\b', text):
            return "مخلوط"  # Mixed sentiment if combined with separation.
        return "مثبت"

    # Rule 4: Check for philosophical or contemplative cues.
    if re.search(r'\bسوال\b', text) or re.search(r'\bخیال\b', text) or re.search(r'\bفلسفہ\b', text):
        return "فلسفیانہ"

    # Fallback: If no strong indicators are present.
    return "غیر جانب دار"

# Step 4: Read the uploaded CSV file.
try:
    df = pd.read_csv('bal-e-jibreel.csv')
except Exception as e:
    print("Error reading the CSV file:", e)
    raise

# Step 5: Ensure the CSV has a "Verse" column.
if 'Verse' not in df.columns:
    print("Error: The CSV file does not contain a 'Verse' column. Please verify the column name.")
else:
    # Apply the sentiment labeling function to each entry in the 'Verse' column.
    df['Sentiment'] = df['Verse'].apply(label_sentiment)

    # Step 6: Save the updated DataFrame to a new CSV file.
    output_filename = 'bal-e-jibreel_with_sentiment.csv'
    df.to_csv(output_filename, index=False, encoding='utf-8-sig')
    print("Sentiment labeling complete. Output saved as '{}'.".format(output_filename))

    # Step 7: Download the output file.
    files.download(output_filename)
```
